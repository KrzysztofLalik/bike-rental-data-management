## Bike Rental Data Cleaning and Analysis

This project involves importing, merging, cleaning, and analyzing monthly Citi Bike rental data from 2016. The goal is to prepare the dataset for downstream insights and identify any patterns or issues in the raw data.

### Library and Data imports

```python
import pandas as pd
import glob
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step: Uploading the data from multiple csv files to one dateframe
The Citi Bike data is stored in monthly files. Let's read those in and concatenate them in a single dataframe.

```python
# Get all CSV file names matching the pattern for 2016 Citi Bike data
csv_files = glob.glob("JC-2016*-citibike-tripdata.csv")

# Load and merge all CSVs into one DataFrame
df = pd.concat((pd.read_csv(f) for f in csv_files), ignore_index = True)
```

### Step: Formatting of the column names

```python
#Housekeeping to remove spaces and uppercase letters
df.columns = [x.replace(' ', '_').lower() for x in df.columns]
```

### Step: Checking Column Data Type Accuracy

```python
df.dtypes
```

### observation: 
- time-related columns are of object type instead of appropriate datetime type,
- naming columns (ex.Start Station Name) could be more useful for pandas analysys in string then object type
- User Type is object which is ok, but category type seems more adequate - can save memory and speed up certain operations
- birth year is float64 intead of int64 probably due to the missing values,
- for the clarity we can embody encoding the Gender column and converting it then to the categorytype 

### Step: Converting column types and encoding gender

```python
#Converting time-related columns to datetime type
df['start_time'] = pd.to_datetime(df['start_time'])
df['stop_time'] = pd.to_datetime(df['stop_time'])

#Converting naming columns to string type 
df['start_station_name'] = df['start_station_name'].astype('string')
df['end_station_name'] = df['end_station_name'].astype('string')

#Converting columns with categories to category type
df['user_type'] = df['user_type'].astype('category')

#Encoding the Gender column and converting it to the category type
gender_map = {0: 'Unknown', 1: 'Male', 2: 'Female'}
df['gender'] = df['gender'].map(gender_map).astype('category')
```

```python
df.dtypes
```

### Step: Validation of declared age 

```python
#Creation of the age column based on the birth date 
df['age'] = 2016 - df['birth_year']
df['age'].sort_values(ascending = False)[:10]
```

```python
#removing faulty data
df.drop([215055], inplace=True)
```

### Step: `trip_duration` validation


```python
#Longest trips in days
df['trip_duration_days'] = df['trip_duration']/(3600*24)
df['trip_duration_days'].sort_values(ascending = False)[:20]
```

```python
#Shortest trips in seconds
df['trip_duration'].sort_values(ascending = True)[:20]
```

### observation:
- The minimum trip duration in the dataset is 61 seconds, which aligns with the data dictionary note that trips under 60 seconds were considered false starts and have already been removed.
- The maximum trip duration is 189 days, which significantly exceeds Citi Bike’s official policy limit of 24 hours (24*3600=86400 seconds). While these long trips may reflect system errors or policy violations (e.g., users not docking bikes properly), we will not remove them. Instead, we’ll add a flag so the analytics team can easily identify and filter them for further investigation if needed.

```python
# create a column to flag rides that do not comply with system protocols
df['valid_duration'] = df['trip_duration'].apply(lambda x: 0 if x > 86400 else 1).astype(bool)
```

### Missing data in `user_type` over the year:

```python
max(df.start_time)
```

```python
min(df.start_time)
```

```python
# Create a month column for grouping
df['month'] = df['start_time'].dt.to_period('M').dt.to_timestamp()

# Create a list of all months in the full range
all_months = pd.date_range(df['month'].min(), df['month'].max(), freq='MS')

# Count alldata and missing
monthly_alldata = df.groupby('month').size()
monthly_missing = df[df['user_type'].isnull()].groupby('month').size()

# Reindex to include all 12 months, filling with 0 where needed
monthly_alladata = monthly_alldata.reindex(all_months, fill_value=0)
monthly_missing = monthly_missing.reindex(all_months, fill_value=0)

# Plot 1: All records - user_type
plt.figure(figsize=(12, 4))
monthly_alldata.plot.bar(color='skyblue', title='user_type - all records')
plt.xlabel('Month')
plt.ylabel('Records - user_type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Missing - user_type
plt.figure(figsize=(12, 4))
monthly_missing.plot.bar(color='salmon', title='user_type - missing records')
plt.xlabel('Month')
plt.ylabel('Missing records - user_type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### observation:
- Although missing values represent a small portion of the total, their distribution suggests they may not be missing at random. From a data management standpoint, we’ll retain these records but replace NaNs with an 'Unknown' category. This allows the analytics team to investigate further and ensures consistent tracking if the dataset is expanded with additional years.

```python
if isinstance(df['user_type'].dtype, pd.CategoricalDtype):
    if 'Unknown' not in df['user_type'].cat.categories:
        df['user_type'] = df['user_type'].cat.add_categories('Unknown')

df['user_type'] = df['user_type'].fillna('Unknown')
```
