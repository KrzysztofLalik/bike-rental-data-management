## Bike Rental Data Cleaning and Analysis

This project involves importing, merging, cleaning, and analyzing monthly Citi Bike rental data from 2016. The goal is to prepare the dataset for downstream insights and identify any patterns or issues in the raw data.

### Library and Data imports

```python
import pandas as pd
import glob

import datetime
import string

import matplotlib.pyplot as plt
import numpy as np

from sqlalchemy import create_engine
```

### Step: Uploading the data from multiple csv files to one dateframe
The Citi Bike data is stored in monthly files. Let's read those in and concatenate them in a single dataframe.

```python
# Get all CSV file names matching the pattern for 2016 Citi Bike data
csv_files = glob.glob("JC-2016*-citibike-tripdata.csv")

# Load and merge all CSVs into one DataFrame
df = pd.concat((pd.read_csv(f) for f in csv_files), ignore_index = True)
```

### Step: Converting column types and encoding gender

```python
#Converting time-related columns to datetime type
df['start_time'] = pd.to_datetime(df['start_time'])
df['stop_time'] = pd.to_datetime(df['stop_time'])

#Converting naming columns to string type 
df['start_station_name'] = df['start_station_name'].astype('string')
df['end_station_name'] = df['end_station_name'].astype('string')

#Converting columns with categories to category type
df['user_type'] = df['user_type'].astype('category')

#Encoding the Gender column and converting it to the category type
gender_map = {0: 'Unknown', 1: 'Male', 2: 'Female'}
df['gender'] = df['gender'].map(gender_map).astype('category')
```

- The maximum trip duration is 189 days, which significantly exceeds Citi Bike’s official policy limit of 24 hours (24*3600=86400 seconds). While these long trips may reflect system errors or policy violations (e.g., users not docking bikes properly), we will not remove them. Instead, we’ll add a flag so the analytics team can easily identify and filter them for further investigation if needed.

```python
# create a column to flag rides that do not comply with system protocols
df['valid_duration'] = df['trip_duration'].apply(lambda x: 0 if x > 86400 else 1).astype(bool)
```

# Postgres database tables

Creating a database schema:
1) we will need the id column for the trips
2) for the clarity we decided to create the columns with the trip duration in minutes and hours too for the analytics team
3) we don't have a possibility to track the direct distance made by bikes to later filter the ones with the highest mileage for the service, but based on the start and end stations longitude and latitude we can at least determine the minimum distance they traversed - later by taking into account, for example, an estimated value of the distance traveled by trip duration, we may be able to identify the most heavily used bikes for the service purposes.
4) Due to the importance of dates for linking weather and trip data, we will create a separate column for the date in the trips table and generate a new table that breaks down the data by day of the week, includes weekend indicators, and assigns each date to a financial quarter.

  (Own notes - datebase schema)

```python
#1. Creation of the id column
df['id'] = df.index
```

```python
#2. Creation of duration of the trips in minutes and hours
df['trip_duration_min'] = (df['trip_duration']/60).round(2)
df['trip_duration_hours'] = (df['trip_duration']/3600).round(2)
```

```python
#How many stations are there in datebase
df['start_station_name'].unique()
```

```python
#3. Using the Haversine formula for creating the function to calculate the distances between the stations based on their logitude and latitude
def haversine_vectorized(lat1, lon1, lat2, lon2):
    R = 3958.8  # Earth radius in miles

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    delta_lat = lat2 - lat1
    delta_lon = lon2 - lon1

    a = np.sin(delta_lat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(delta_lon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    return (R * c).round(2)

#Calling Haversine formula on our df
df['distance'] = haversine_vectorized(df['start_station_latitude'], df['start_station_longitude'], df['end_station_latitude'], df['end_station_longitude'])
df.distance.head
```

```python
(df['distance'] == 0).sum()
```

```python
#Flagging records with returning bikes to the same station (distance = 0 miles) for the analytics 
df['distance_validation'] = (df['distance'] == 0)
```

## `dates`


```python
#Creation of the dates dataframe

#Range of dates from January 1st to December 31st:
date = pd.date_range(start='2016-01-01', end='2016-12-31', freq='D')

df_dates = pd.DataFrame({'full_date': date})

print(df_dates)
```

```python
# generate numeric month
df_dates['month'] = df_dates['full_date'].apply(lambda x: x.month)
# generate numeric day
df_dates['day'] = df_dates['full_date'].apply(lambda x: x.day)
# generate name of the month
df_dates['month_name'] = df_dates['full_date'].apply(lambda x: x.strftime('%B'))
# generate name of the day
df_dates['day_name'] = df_dates['full_date'].apply(lambda x: x.strftime('%A'))
# generate true of false whether it's weekend
df_dates['weekend'] = df_dates['day_name'].apply(lambda x: 1 if (x == 'Saturday' or x=='Sunday') else 0).astype(bool)
# generate financial quarter
df_dates['quarter'] = df_dates['full_date'].dt.quarter
# generate date_key to correspond with other tables
df_dates['date_key'] = df_dates['full_date'].astype(str).str.replace("-", "")
df_dates.head()
```
